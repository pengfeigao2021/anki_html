# distance definitions

```python
import tqdm

def binary_distance(df_in, col, i, j):
    v1 = df_in.iloc[i][col]
    v2 = df_in.iloc[j][col]
    if v1 == v2:
        return 0
    else:
        return 1
    
def abs_norm_distance(df_in, col, i, j):
    gap = df_in[col].max() - df_in[col].min()
    v1 = df_in.iloc[i][col]
    v2 = df_in.iloc[j][col]
    return np.abs(v1 - v2) / gap

def distance(df_in, col, i, j, method="binary"):
    if method == "binary":
        return binary_distance(df_in, col, i, j)
    else:
        return 1

def get_distance_matrix(df_in):
    n = df_in.shape[0]
    m = df_in.shape[1]
    
    dmat = np.zeros((n, n), dtype=np.float32)
    min_dist = m
    for i in tqdm.tqdm(range(n), total=n):
        for j in range(i + 1, n):
            d = 0
            for col in df_in.columns:
                d += distance(df_in, col, i, j)
            dmat[i,j] = d
            dmat[j,i] = d
            min_dist = min(d, min_dist)

    print('minimum distance: {}'.format(min_dist))
    return dmat
    
dmat = get_distance_matrix(training_df.head(n=100))
print(dmat)
display(dmat)
```

```python
from sklearn.datasets import make_blobs
from sklearn.metrics import pairwise_distances
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np


# Apply DBSCAN with precomputed distance matrix
dbscan = DBSCAN(eps=7, min_samples=2, metric='precomputed')
labels = dbscan.fit_predict(dmat)
labels
```

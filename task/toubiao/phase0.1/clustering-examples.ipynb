# clustering examples

DBSCAN



```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data (2 clusters, plus some noise)
X, _ = make_blobs(n_samples=300, centers=2, cluster_std=0.60, random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Sample Data")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Create a DBSCAN instance
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit and predict clusters
labels = dbscan.fit_predict(X)

print("Cluster labels for each point:", labels)
```

```python
import numpy as np

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma', s=30)
plt.title("DBSCAN Clustering Result")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

# Example: DBSCAN with Precomputed Distance Matrix

```python
from sklearn.datasets import make_blobs
from sklearn.metrics import pairwise_distances
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=0.50, random_state=0)

# Compute pairwise Euclidean distance matrix
D = pairwise_distances(X, metric='euclidean')
```

```python
# print info
print(X.shape)
print(D.shape)
```

```python
# Apply DBSCAN with precomputed distance matrix
dbscan = DBSCAN(eps=1.5, min_samples=5, metric='precomputed')
labels = dbscan.fit_predict(D)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma', s=40)
plt.title("DBSCAN with Precomputed Distance Matrix")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
